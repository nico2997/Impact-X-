```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("PIK data gases.csv")
data
```

```python
x = plt.figure(figsize=(12,6))
plt.plot(range(1,13),data.N2O, marker='o', linewidth=3.0, label = 'Nitrous Oxide (N2O)')
plt.plot(range(1,13),data.CH4, marker='o', linewidth=3.0, c='red', label = 'Methane (CH4)')
plt.xticks(range(1,13),data.Year)
plt.title("China CH4 and N2O Produced from Agriculture Sector in 2007 to 2018", fontsize=18)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO2 equivalent in Mt (Million tons)', fontsize=12)
plt.axes().xaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper left')
for i, v in enumerate(data.N2O):
    if i==2:
        plt.text(i+1, v-4.3, "%r" %v, ha="center")
    else:
        plt.text(i+1, v+2, "%r" %v, ha="center")

for i, v in enumerate(data.CH4):
    if i==3:
        plt.text(i+1, v-4.3, "%r" %v, ha="center")
    else:
        plt.text(i+1, v+2, "%r" %v, ha="center")
plt.ylim(300, 401)


plt.show()
x.savefig('gas1.png')
```

```python
y = plt.figure(figsize=(12,6))
plt.plot(range(1,13),data.CO2, marker='o', linewidth=3.0, c='green', label = 'Carbon Dioxide (CO2)')
plt.xticks(range(1,13),data.Year)
plt.title("China CO2 Produced from Agriculture Sector in 2007 to 2018", fontsize=18)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO2 equivalent in Mt (Million tons)', fontsize=12)
plt.axes().yaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper left')

for i, v in enumerate(data.CO2):
        plt.text(i+1, v+0.8, "%r" %v, ha="center")
        
plt.yticks(np.arange(0, 21, 3))

plt.show()
y.savefig('gas2.png')
```
