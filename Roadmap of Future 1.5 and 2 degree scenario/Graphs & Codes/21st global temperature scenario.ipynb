```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("1.5.csv")
data
```

```python
x = plt.figure(figsize=(10,8))
plt.plot(range(1,102),data.Historical, linewidth=3.0, c='black', label = 'Historic')
plt.plot(range(1,102),data.y2010, linewidth=1, c='maroon', label = '2010')
plt.plot(range(1,102),data.y2011, linewidth=1, c='red', label = '2011')
plt.plot(range(1,102),data.y2012, linewidth=1, c='orangered', label = '2012')
plt.plot(range(1,102),data.y2013, linewidth=1, c='salmon', label = '2013')
plt.plot(range(1,102),data.y2014, linewidth=1, c='salmon', label = '2014')
plt.plot(range(1,102),data.y2015, linewidth=1, c='papayawhip', label = '2015')
plt.plot(range(1,102),data.y2016, linewidth=1.0, c='peachpuff', label = '2016')
plt.plot(range(1,102),data.y2017, linewidth=1.0, c='mistyrose', label = '2017')
plt.plot(range(1,102),data.y2018, linewidth=1.0, c='peachpuff', label = '2018')
plt.plot(range(1,102),data.y2019, linewidth=1, c='dimGray', label = '2019')
plt.plot(range(1,102),data.y2020, linewidth=1.0, c='lavender', label = '2020')
plt.plot(range(1,102),data.y2021, linewidth=1.0, c='powderblue', label = '2021')
plt.plot(range(1,102),data.y2022, linewidth=1.0, c='lightskyblue', label = '2022')
plt.plot(range(1,102),data.y2023, linewidth=1.0, c='deepskyblue', label = '2023')
plt.plot(range(1,102),data.y2024, linewidth=1.0, c='royalblue', label = '2024')
plt.plot(range(1,102),data.y2025, linewidth=1.0, c='blue', label = '2025')
plt.plot(range(1,102),data.y2026, linewidth=1.0, c='darkBlue', label = '2026')
plt.xticks(range(1,102,10),[2000,2010,2020,2030,2040,2050,2060,2070,2080,2090,2100])
plt.xlabel('Year', fontsize=12)
plt.ylabel('$\mathregular{CO^{2}}$ emissions from fossil fuels\nand land-use (Gt $\mathregular{CO^{2}}$/yr)', fontsize=12, multialignment='center')
plt.axes().yaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper right')


        
plt.yticks(np.arange(0, 50,5))

plt.show()
x.savefig('1.5degree.png')
```
