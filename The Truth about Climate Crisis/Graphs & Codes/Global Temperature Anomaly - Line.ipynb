```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Land Ocean Temperature.csv")
data
```

```python
x=plt.figure(figsize=(12,6))
plt.plot(range(1,52),data.Real,marker='o', linewidth=2.0, label = 'Annual Mean Temperature', color = 'gray')
plt.plot(range(1,52),data.Temperature, linewidth=2.0, label = 'Smoothing Version', color = 'black')
plt.xticks(range(1,52,5),[1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2020])
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature Anomaly (C)', fontsize=12)
plt.axes().yaxis.grid(c='black',alpha=0.2)
plt.axes().xaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper left')
plt.axhline(y=0, color='red', linestyle='-', lw=1.5)

plt.yticks(np.arange(-0.3,1.2,0.2))
#plt.plot([51], [1.01], 'o')
plt.text(51.3, 0.96, '1.02', fontsize=11,  color='black')

plt.annotate('First Oil Crisis (1973)', xy=(4, 0.17), xytext=(0.01, 0.4), fontsize=10,
            arrowprops=dict(facecolor='black', shrink=0.1))

#plt.annotate('2nd Oil Crisis (1979)', xy=(10.5, 0.18), xytext=(6.2, 0.45), fontsize=10,
#            arrowprops=dict(facecolor='black', shrink=0.1))

plt.text(37.5, 0.05, '1951-1980 average temperatures', fontsize=11,  color='red')

plt.show()
x.savefig('Temperature anomaly.png')
```
