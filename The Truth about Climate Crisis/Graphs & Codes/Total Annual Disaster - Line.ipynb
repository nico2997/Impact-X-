```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Natural Disaster 1970-2020.csv")
data
```

```python
x = plt.figure(figsize=(12,6))
plt.plot(range(1,52),data.Cases, linewidth=2.0, label = 'Total Annual Disaster')
plt.xticks(range(1,52,5),[1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2020])
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Events', fontsize=12)
plt.axes().yaxis.grid(c='black',alpha=0.2)
plt.axes().xaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper left')

plt.plot([27], [227], 'o',color='k')
plt.text(26.2, 207, '227', fontsize=11,  color='black')
plt.plot([31], [411], 'o',color='black')
plt.text(30.2, 421, '411', fontsize=11,  color='black')

plt.plot([35.9], [433], 'o')

plt.annotate('Peak', xy=(36.5, 433), xytext=(38.5, 430), fontsize=11,
            arrowprops=dict(facecolor='black',lw=3, arrowstyle='->'))


plt.plot([51], [389], 'o',color='red')
plt.text(51, 370, '389', fontsize=11,  color='black')

#plt.text(37.5, 0.05, '1951-1980 average temperatures', fontsize=11,  color='red')

plt.show()
x.savefig('Total Disaster.png')
```
