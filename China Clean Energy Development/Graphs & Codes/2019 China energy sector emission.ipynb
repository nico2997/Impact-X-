```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("China CO2 emission by sectors 2019.csv")
data
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:1.1f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
y = plt.figure(figsize=(8,8))
explode = (0.05, 0, 0, 0,0)
plt.pie(data["Percentage"], labels=data.Industry, autopct='%1.0f%%', textprops={'fontsize': 11.8},
         shadow=False, startangle=180)
plt.rcParams['patch.linewidth'] = 1
plt.title("Share of energy related CO2 in China in 2019, by sector", fontsize=20)
plt.show()
y.savefig('chinasector2019.png')
```
