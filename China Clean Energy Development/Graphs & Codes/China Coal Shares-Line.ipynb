```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("China share of coal consumption.csv")
data
```

```python
from matplotlib.ticker import PercentFormatter
y = plt.figure(figsize=(12,6))
plt.plot(range(1,32),data.Coal, linewidth=3.0, c='black')
plt.xticks(range(1,32,5),[1990,1995,2000,2005,2010,2015,2020])
plt.xlabel('Year', fontsize=12)
plt.ylabel('China Coal Shares Consumption (Percentage)', fontsize=12)
plt.axes().yaxis.grid(c='black',alpha=0.2)
plt.axes().xaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper right')

plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))
plt.yticks(np.arange(50, 90, 5))

plt.show()
y.savefig('coal.png')
```
