```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("World GHG 2018.csv")
data
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:1.1f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
y = plt.figure(figsize=(8,8))
explode = (0.05, 0, 0, 0,0)
plt.pie(data["Total CO2e"], labels=data.Sector, autopct='%1.1f%%', textprops={'fontsize': 11.8},
        explode = explode, shadow=True, startangle=180)
plt.rcParams['patch.linewidth'] = 1
plt.title("The Percentage of World CO2e Produced by Each Sector in 2018", fontsize=20)
plt.show()
y.savefig('world2018GHG.png')
```

```python
sns.set()
data["Total CO2e"].plot(kind='pie',labels=data.Sector, figsize=[10,10],
          autopct= '%1.1f%%', textprops={'fontsize': 12.8}, shadow=True, startangle=160)
plt.title("The Percentage of Primary Plastic Production by Industrial Sector in 2015", fontsize=20)
plt.show()
```
