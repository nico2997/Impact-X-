```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Energy GHG 2018.csv")
data
```

```python
y = plt.figure(figsize=(10,10))
explode = (0, 0, 0, 0,0)
plt.pie(data["Total CO2e"], labels=data.Sector, autopct='%1.1f%%', textprops={'fontsize': 12.8}, startangle=0)
plt.rcParams['patch.linewidth'] = 2
plt.show()
y.savefig('energy2018GHG.png')
```
