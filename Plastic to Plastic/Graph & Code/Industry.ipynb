```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Industry.csv")
data
```

```python
y = plt.figure(figsize=(12,12))
_,_,autopcts = plt.pie(data.Production, explode =(0.05, 0.05, 0.05, 0, 0, 0,0,0), 
                       labels=data.Industry, autopct='%1.1f%%', textprops={'fontsize': 11.8})
plt.title("The Percentage of Primary Plastic Production by Industrial Sector in 2015", fontsize=20)
plt.setp(autopcts, **{'color':'black', 'weight':'bold', 'fontsize':12.5})
plt.show()
y.savefig('plastic2.png')
```

```python
z=plt.figure(figsize=(23,7))
clrs = ['blue' if (x<22) else 'grey' if (x < 80) else 'red' for x in data.dec ]

plt.barh(data.Industry,data.dec, color =clrs)
for i, v in enumerate(data.dec):
    plt.text(v+0.5 , i, f"{v}%", color='black', fontsize=13, va='center',weight='bold')

plt.title("Percentage of Plastic Waste Generation by Industrial Sector in 2015", fontsize=20)
plt.xlabel('Percentage', fontsize=14)
plt.axes().xaxis.grid(c='black',alpha=0.2)
plt.show()
z.savefig('plastic3.png')
```
