```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Cumulative.csv")
data
```

```python
data_pred = data[7:9]
data_pred
```

```python
x = plt.figure(figsize=(12,6))
plt.plot(range(1,10),data.Production, marker='o', linewidth=3.0, label = 'Actual 1950-2020')
plt.plot(range(8,10),data_pred.Production, marker='o', linewidth=3.0, c='red', label = '2050 prediction')
plt.xticks(range(1,10),data.Year)
plt.title("Global cumulative production of plastic 1950-2050 (in million metric tons)", fontsize=18)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Production Volume in million Metric Tons', fontsize=12)
plt.axes().yaxis.grid(c='black',alpha=0.2)
plt.legend(loc='upper left')
for i, v in enumerate(data.Production):
    if i==0:
        plt.text(i+1, v+1000, "2 million", ha="center")
    elif i==1:
        plt.text(i+1, v+1000, "47 million", ha="center")
    elif i==2:
        plt.text(i+1, v+1000, "249 million", ha="center")
    elif i==3:
        plt.text(i+1, v+1000, "798 million", ha="center")
    elif i==4:
        plt.text(i+1, v+1000, "1740 million", ha="center")
    elif i==5:
        plt.text(i+1, v+1500, "3390 million", ha="center")
    elif i==6:
        plt.text(i+1, v+1500, "6060 million", ha="center")
    elif i==7:
        plt.text(i+1.5, v+1, "9334 million", ha="center")
    elif i==8:
        plt.text(i+0.9, v+1000, "34000 million", ha="center")
plt.ylim(0, 40000)

plt.show()
x.savefig('plastic5.png')
```
