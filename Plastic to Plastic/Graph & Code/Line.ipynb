```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Graph.csv")
data
```

```python
data_1 = data[1:17]
data_1
```

```python
data_pred = data[16:18]
data_pred
```

```python
x = plt.figure(figsize=(12,6))
plt.plot(range(1,19),data.Waste, marker='o', linewidth=3.0, label = 'Actual 1950-2020')
plt.plot(range(17,19),data_pred.Waste, marker='o', linewidth=3.0, c='red', label = '2040 prediction')
plt.xticks(range(1,19),data.Year)
plt.title("Global Plastic Production from 1950 to 2020 (in million metric tons)", fontsize=18)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Production Volume in Million Metric Tons', fontsize=12)
plt.grid(c='black',alpha=0.2)
plt.legend(loc='upper left')
for i, v in enumerate(data.Waste):
    if i==0:
        plt.text(i+0.89, v+25, "%r" %v, ha="center")
    else:
        plt.text(i+0.89, v+25, "%d" %v, ha="center")
plt.ylim(0, 600)

plt.show()
x.savefig('plastic1.png')
```
